### Barrhaven Data, 3 km radius

```python
#STEP 1) create function to call data
```

```python
def fsq_API_call(lat, long, radius, categories, key):
    '''
    returns a normalized json of the data
    '''
    url = "https://api.foursquare.com/v3/places/search?ll={}%2C{}&radius={}&categories={}&limit=50&fields=rating%2Cname%2Cdistance%2Ccategories%2Cfsq_id".format(lat,long,radius,categories)
    headers = {"accept": "application/json","Authorization": key}
    response = requests.get(url, headers=headers)
    data = response.json()
    #print(data)
    return data
```

```python
barr_bars = fsq_API_call("45.27368231718483", "-75.75939435338674", "3000","13003", fs_api_key)
barr_ae = fsq_API_call("45.27368231718483", "-75.75939435338674", "3000","10000", fs_api_key)
barr_restau = fsq_API_call("45.27368231718483", "-75.75939435338674", "3000","13065", fs_api_key)
barr_land = fsq_API_call("45.27368231718483", "-75.75939435338674", "3000","16000", fs_api_key)
barr_other = fsq_API_call("45.27368231718483", "-75.75939435338674", "3000","13028,13029,13040,13059,13032", fs_api_key)
```

```python
#STEP 2) normalize and combine data
#I didn't do this within the function above because I need the raw data to normalize the categories
```

```python
barr_bars_norm = json_normalize(barr_bars['results'])
barr_ae_norm = json_normalize(barr_ae['results'])
barr_restau_norm = json_normalize(barr_restau['results'])
barr_land_norm = json_normalize(barr_land['results'])
barr_other_norm = json_normalize(barr_other['results'])
```

# Yelp

Send a request with the same location paramaters (location, radius, etc)

```python
yelp_api_key = os.environ['yelp_api_key']
yelp_id = os.environ["yelp_client_id"]
```

```python
#set the Authorization HTTP header value as Bearer API_KEY.
```

```python
#ZERO RESULTS FOR SENEGAL
#use stone house: 45.27368231718483, -75.75939435338674
```

```python
def yelp_API_call(lat, long, radius, categories, key):
    '''
    returns a normalized json of the data
    '''
    headers = {"accept": "application/json",
           "Authorization": f"BEARER {key}"}

    param = {"latitude" : lat,
             "longitude" : long,
             "radius" : radius,
             "limit" : "50",
             "categories" : categories}
    url = "https://api.yelp.com/v3/businesses/search?"
    response = requests.get(url=url, headers=headers, params = param)
    data= response.json()
    return data
```

```python
#https://www.yelp.com/developers/documentation/v3/all_category_list
yelp_arts = yelp_API_call("45.27368231718483", "-75.75939435338674", "3000", "arts", yelp_api_key)
yelp_land = yelp_API_call("45.27368231718483", "-75.75939435338674", "3000", "landmarks", yelp_api_key)
```

```python
yelp_food = yelp_API_call("45.27368231718483", "-75.75939435338674", "3000", "food", yelp_api_key)
yelp_restau = yelp_API_call("45.27368231718483", "-75.75939435338674", "3000", "restaurants", yelp_api_key)
```

```python
def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection
```
